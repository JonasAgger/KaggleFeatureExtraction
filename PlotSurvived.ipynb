```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
train = pd.read_csv("train.csv")

age_survived = train.loc[train['Survived']==1, "Age"]
age_survived.plot.hist(fc=(0, 0, 1, 0.5), normed=1, label="Survived")
age_did_not_survive = train.loc[train['Survived']==0, "Age"]
age_did_not_survive.plot.hist(fc=(1, 0, 0, 0.5), normed=1, label="Did not Survive")
plt.xlabel("Age")
plt.ylabel("Percentage of Passengers")
plt.legend(loc='upper right')
plt.title("Distribution of Age of Survivors and Non-Survivors")
plt.plot()
```

```python
embarked_survived = train.loc[train['Survived']==1, "Embarked"]
embarked_did_not_survive = train.loc[train['Survived']==0, "Embarked"]
embarked_did_not_survive.value_counts().plot(kind="bar")
plt.xlabel("Embarked")
plt.ylabel("Number of Passengers")
plt.legend(loc='upper right')
plt.title("Distribution of Embarkment of Non-Survivors")
plt.show()


embarked_survived.value_counts().plot(kind="bar")

plt.xlabel("Embarked")
plt.ylabel("Number of Passengers")
plt.legend(loc='upper right')
plt.title("Distribution of Embarkment of Survivors")
```

```python
pclass_survived = train.loc[train['Survived']==1, "Pclass"]
pclass_survived.plot.hist(fc=(0, 0, 1, 0.5), normed=1, label="Survived")
pclass_did_not_survive = train.loc[train['Survived']==0, "Pclass"]
pclass_did_not_survive.plot.hist(fc=(1, 0, 0, 0.5), normed=1, label="Did not Survive")
plt.xlabel("PassengerClass")
plt.ylabel("Percentage of Passengers")
plt.legend(loc='upper right')
plt.title("Distribution of PassengerClass of Survivors and Non-Survivors")
plt.plot()
```
